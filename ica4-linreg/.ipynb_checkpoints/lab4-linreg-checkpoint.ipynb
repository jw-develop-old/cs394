## 2. k-neighbors regression

Before doing linear regression, here's an interesting point of comparison. KNN for *classificaiton* should be fresh in your mind, but it turns out we can use a similar strategy for regression as well. For any new x value, the model returns a target value equal to the average target value of whatever training data points are closest to x. Consider this on the wave dataset with three test inputs, using one neighbor (the function `plot_knn_regression` does all of the wor here, including showing the plot):

```python
mglearn.plots.plot_knn_regression(n_neighbors=1)
```

...and with three neighbors:

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

Let's do this more carefully and see how it performs. The `KNeighborsRegressor` class from `sklearn.neighbors` fits a model to training data, which can then be used to make predictions.  Keep in mind, though, that "fitting a model" in this case means simply storing the training data. As with a KNN classifier, the work is put off until a prediction is done.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)
# fit the model using the training data and training targets
reg.fit(X_train, y_train)
```

Now let's runt it on the test set.

```python
print("Test set predictions:\n", reg.predict(X_test))
```

Well, how good is that? We can use the "r-squared score" (formally called the *coefficient of determination*), which is a measure of how well two variables are correlated---in this case, how well the predictions match the known targets in the test data. This score ranges from 0 (unrelated) to  1 (perfect correlation). 

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

A score of 0.83 is "not bad." But let's try to visualize this. The model is a real-valued function, so we can plot it. Let's see what it looks like with various settings for the
neighbors parameter, and plot the training and test data with it.

```python
import numpy as np
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
                "Test data/target"], loc="best")
```

## 3. Simple linear regression

Now we turn our attention to linear models for regression, as we've been studying in class.
Taking the wave dataset, split into training and test sets, and fit using `LinearRegression` in `sklearn`:

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

This is just a line in the plane, defined by a slope and y-intercept:

```python
print("lr.coef_:", lr.coef_)
print("lr.intercept_:", lr.intercept_)
```

But let's plot it and see how the line looks in the contex of the training data.

```python
from mglearn.plot_helpers import cm2

line = np.linspace(-3, 3, 100).reshape(-1, 1)
plt.figure(figsize=(8, 8))
plt.plot(line, lr.predict(line))
plt.plot(X, y, 'o', c=cm2(0))
ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')
ax.set_ylim(-3, 3)
ax.legend(["model", "training data"], loc="best")
ax.grid(True)
ax.set_aspect('equal')
```

Test your own understanding at this point. The line is fit so as to minimize the average squared distance between points and the line. This approach is very simple, with only the slope and intercept as parameters.

And we can test this model formally:

```python
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test, y_test)))
```

That is quite a bit lower than the scores for KNN. Notice in particular that the score on the *training set* isn't that good eaither. This simple model is underfitted.

But let's see how things look in a more complex data set. Remember that the Extended Boston dataset has 105 features. We can't visualize this (unless you know of a way to make a 105-dimensional graph), but we can still compute scores for it. This model is in the form of 

\begin{equation}
y(\mathbf{x}) = w_0 + \sum_{i=1}^{105} w_i x_i
\end{equation}

Where $x_i$ is a product of two features.  

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test, y_test)))
```

The score on the test set is even worse, but notice the training set is highly correlated to the model. This is a case of severe overfitting.
This leads us to alternatives to simple linear regression that make use of 
*regularization*. This means that we add terms to the equation that restrict
the model

## 4.  Ridge regression

Ridge regression---or, ridge regularization---restricts the model by penalizing the Euclidean distance of the coefficients, or generally of the parameter array $\mathbf{w}$. Instead of finidng the parameters that minimize the average square of distance from data points to the linear-regression

\begin{equation}
||A \mathbf{x} - b||_2
\end{equation}

...it chooses parameters that minimize the sum of that and some scaling of the identity matrix


\begin{equation}
||A \mathbf{x} - b||_2 + || \alpha I||_2
\end{equation}

The meta-parameter $\alpha$ determines how much weight to give to this additional term.

Here it is on the Extended Boston Housing data from before:


```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test, y_test)))
```

The upshot is that is that the resulting model is not as fit to the training data, but the performance on the test set has improved noticably. The $\alpha$ parameter is a lever for controlling how restricted the model is. Try increase or decreasing it:

```python
ridge10 = Ridge(alpha=1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

What does this suggest?  Remember that a model's performance on new data is called *generalization.*

Let's visualize how $\alpha$ affects the weight vector, which is the `coef_` attribute.

```python
plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")

plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.legend()
```

Higher values of $\alpha$ result in a smaller magnitude for the coefficients. As M&uuml;ller and Guido put it,

> The x-axis enumerates the entries of `coef_`. Thus `x=0` shows the coefficient associated 
> with the first feature, `x=1` the coefficient associated with the second feature, up to >`x=100`. The y-axis shows the numeric values of the corresponding values of the coefficients. >The main takeaway here is that for `alpha=10`, the coefficients are mostly between around -3 >and 3. The coefficients for the `Ridge` model with `alpha=1` are somewhat larger. The dots >corresponding to `alpha=0.1` have larger magnitude still, and many of the dots corresponding >to linear regression without any regularization (which would be `alpha=0`) are so large they >are outside of the chart. (Pg 53)

Now consider fixing $\alpha$ at 1 and plot the size of the training set against the r-squared score. We'll throw in linear regression for comparison. 

```python
from sklearn.model_selection import learning_curve, KFold

def plot_learning_curve(est, X, y):
    training_set_size, train_scores, test_scores = learning_curve(
        est, X, y, train_sizes=np.linspace(.1, 1, 10), cv=KFold(20, shuffle=True, random_state=1))
    estimator_name = est.__class__.__name__
    line = plt.plot(training_set_size, train_scores.mean(axis=1), '--',
                    label="training " + estimator_name)
    plt.plot(training_set_size, test_scores.mean(axis=1), '-',
             label="test " + estimator_name, c=line[0].get_color())
    plt.xlabel('Training set size')
    plt.ylabel('Score (R^2)')
    plt.ylim(0, 1.1)

X, y = mglearn.datasets.load_extended_boston()

plot_learning_curve(Ridge(alpha=1), X, y)
plot_learning_curve(LinearRegression(), X, y)
plt.legend(loc=(0, 1.05), ncol=2, fontsize=11)
```

## 5. Lasso

An alternative to regression with ridge regularization is *lasso*. The name comes from *least absolute shrinkage and selection operator*. The difference from ridge is that it uses the L1 norm intead of L2---that is, it uses Manhattan distance instead of Euclidean distance. As a result, some coefficients are zero, and the corresponding attributes are ignored.

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

Okay, that was bad. By far most of the 105 features were ignored. But as with ridge, we can fiddle with $\alpha$. We also need to adjust a parameter called `max_iter` which I haven't figured out yet, appart from that it obviously has something to do with the number of iterations an underlying algorithm is doing.

```python
# we increase the default setting of "max_iter",
# otherwise the model would warn us that we should increase max_iter.
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso001.coef_ != 0))
```

That's better. What if we set $\alpha$ really low?

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso00001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso00001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso00001.coef_ != 0))
```

That's too many features, and overfitting. Let's compare the coeffient magnitudes for different values of $\alpha$, with the best ridge model ($\alpha = 0.1$) for another point of comparison:

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.0001")

plt.plot(ridge01.coef_, 'o', label="Ridge alpha=0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
```
